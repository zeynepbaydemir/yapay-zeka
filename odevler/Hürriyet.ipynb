### Web Scraping


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://www.hurriyet.com.tr/'
response = requests.get(url, headers = {'User-agent': 'your bot 0.1'})
```

```python
response.text
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup.beautify
```

```python
continer = soup.findAll('div',{'class':'showcase row'})
```

```python
# Haber başlıklarını ve yayın tarihlerini çekelim.
basliklar = soup.findAll('span', {'class':'box__title box-lg-3__title'})
aciklamalar = soup.findAll('p', {'class':'layout5-fullbox__exp'})
tagler = soup.findAll('a', {'class':'box__tag', 'data-google-interstitial':'false',  'target':"_blank"})

print(basliklar)
print(tagler)
print(aciklamalar)
```

```python
b = []
a = []
t = []
columns = ['basliklar', 'aciklamalar','tagler']
for(b1, a1, t1) in zip(basliklar, aciklamalar, tagler):
    b.append(b1.text)
    a.append(a1.text)
    t.append(t1.text)
```

```python
d = {'basliklar':b, 'aciklamalar':a, 'tagler':t}
pd.DataFrame(d)
```

```python
# Her bir elementin metin içeriğini al
basliklar_text = [element.text.strip() for element in basliklar]
aciklamalar_text = [element.text.strip() for element in aciklamalar]
tagler_text = [element.text.strip() for element in tagler]

# Elde ettiğiniz text içeriklerini bir dosyaya yazın
with open('Hürriyet.txt', 'w', encoding='utf-8') as file:
    file.write("Başlıklar:\n")
    for item in basliklar_text:
        file.write("%s\n" % item)

    file.write("\nTag'ler:\n")
    for item in tagler_text:
        file.write("%s\n" % item)

    file.write("\nAçıklamalar:\n")
    for item in aciklamalar_text:
        file.write("%s\n" % item)
```
